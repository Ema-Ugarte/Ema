- `order_products.csv`: cada fila corresponde a un artículo pedido en un pedido.
    - `'order_id'`: número de ID que identifica de manera única cada pedido.
    - `'product_id'`: número ID que identifica de manera única cada producto.
    - `'add_to_cart_order'`: el orden secuencial en el que se añadió cada artículo en el carrito.
    - `'reordered'`: 0 si el cliente nunca ha pedido este producto antes, 1 si lo ha pedido.
- `aisles.csv`
    - `'aisle_id'`: número ID que identifica de manera única cada categoría de pasillo de víveres.
    - `'aisle'`: nombre del pasillo.
- `departments.csv`

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
numero_De_rep = df_order.groupby("product_id")["reordered"].sum()

total_de_pedidos  = df_order.groupby("product_id")["reordered"].count()
```

```python
df_ratio = (numero_De_rep / total_de_pedidos).reset_index()

df_ratio.columns = ["product_id", "order_ratio"]
```

```python
df_final_ratio  = df_ratio.merge(df_products, on= "product_id", how="left")

print(df_final_ratio[["product_id", "product_name","order_ratio"]].sort_values("order_ratio", ascending=False))
```

Escribe aquí tus conclusiones.

### [C4] Para cada cliente, ¿qué proporción de sus productos ya los había pedido?

```python
df = df_instacart.merge(df_order, on='order_id')

numero_de_pedidos = df.groupby("user_id")["reordered"].sum()

total_de_pedidos = df.groupby("user_id")["reordered"].count()

proporcion_de_pedidos = (numero_de_pedidos / total_de_pedidos).reset_index()

proporcion_de_pedidos.columns = ["user_id", "user_ratio"]

proporcion_final = proporcion_de_pedidos.merge(df_instacart, on="user_id")

print(proporcion_final[["user_id", "user_ratio"]].sort_values("user_ratio", ascending=False))
```
